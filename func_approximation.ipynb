```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.linalg import solve
import seaborn as sns
import math
from scipy import linalg
%matplotlib inline
```

```python
def f(x):
    return math.sin(x/5.0) * math.exp(x/10.0) + 5.0 * math.exp(-x / 2.0)
min = 1
max = 15
xx = np.arange(min,max+0.1,0.1)
yy = np.array(list(map(f, xx)))
plt.figure(figsize=(8,6))
plt.plot(xx, yy)
plt.show()
```

```python
X = np.ones((2,2))
X[0, 1] = min
X[1, 1] = max
Y = np.ones((2,1))
Y[0,0] = f(min)
Y[1,0] = f(max)
ans = linalg.solve(X, Y)

yy_ans = ans[0, 0] + xx*ans[1, 0]
plt.figure(figsize=(8,6))
plt.plot(xx, yy)
plt.plot(xx, yy_ans, color='r')
plt.show()
print(ans)
```

```python
X = np.ones((3,3))
X[0, 1] = min
X[0, 2] = min**2
X[1, 1] = 8
X[1, 2] = 8**2
X[2, 1] = max
X[2, 1] = max**2
Y = np.ones((3,1))
Y[0,0] = f(min)
Y[1,0] = f(8)
Y[2,0] = f(max)
ans = linalg.solve(X, Y)

yy_ans = ans[0, 0] + xx*ans[1, 0] + ans[2, 0]*xx**2
plt.figure(figsize=(8,6))
plt.plot(xx, yy)
plt.plot(xx, yy_ans, color='r')
plt.show()
print(ans)
```

```python
X = np.ones((4,4))
Y = np.ones((4,1))
def mnogochlen(znach, stepen, matr, pos, res):
    for i in range(stepen+1):
        X[pos, i] = znach**i
        Y[pos] = f(znach)
dotes = [min, 4, 10, max]
for p, d in enumerate(dotes):
    mnogochlen(d, 3, X, p, Y)

ans = linalg.solve(X, Y)

yy_ans = ans[0, 0] + xx*ans[1, 0] + ans[2, 0]*xx**2 + ans[3, 0]*xx**3
plt.figure(figsize=(8,6))
plt.plot(xx, yy)
plt.plot(xx, yy_ans, color='r')
plt.show()
print(ans)
```

```python
ans_f = 'answer2.txt'
answer = ''
for k in ans:
    answer += str(round(k[0], 2)) + ' '
answer = answer.strip()
print(answer)
with open(ans_f, 'w') as fl:
    fl.write(answer)
```
